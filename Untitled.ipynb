```python
network_data_2022_2023 = pd.read_csv("network_data_2022_2023.csv")
```

```python
G = nx.from_pandas_edgelist(network_data_2022_2023, 'authname1', 'authname2', edge_attr='value', create_using=nx.Graph())
```

```python
pos = nx.kamada_kawai_layout(G)

nx.draw(G, with_labels=True, node_color='skyblue', edge_cmap=plt.cm.Blues, pos=pos)
```

```python
from pyvis.network import Network


net = Network(notebook=True, cdn_resources='remote', width=1400, height=800, bgcolor="white", font_color="black")
net.show_buttons(filter_=['physics'])

node_degree = dict(G.degree)

# Multiply node sizes by two
node_degree_doubled = {node: 3 * degree for node, degree in node_degree.items()}


# Set the node attributes with the doubled sizes
nx.set_node_attributes(G, node_degree_doubled, 'size')

#listnodes = net.nodes
#net.nodes.__getitem__(1)

#listnodes = net.nodes

net.from_nx(G)
net.show("networks/network_2022_2023.html")
```
